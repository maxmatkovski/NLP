```python
import pandas as pd

df = pd.read_csv('FINAL-SENTIMENT-TWEETS.csv')
```

```python
a = df['sentiment'].value_counts()
y = [a[1],a[2],a[3],a[4],a[5]]
x = [1,2,3,4,5]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme()
plt.bar(x,y)

plt.xlabel("100,000 Climate Change Tweets - BERT Sentiment Score")
plt.ylabel("Frequency")
```

```python
percents = []
length = len(df)
for each in y:
    percents.append(each/length)
    
percents
```

```python
sum(percents)
```

```python
def average(l):
    
    length = sum(l)
    diff = 0
    val = 0
    for i in range(len(l)):
        diff = l[i] * (i+1)
        val += diff
        print(val)
    result = val/length
    return result
```
