```python
import pandas as pd

df = pd.read_csv('Data/100kTweetsCleaned.csv')
length = len(df)
df
```

```python
!pip install transformers requests beautifulsoup4 pandas numpy
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch
import requests
import re
```

```python
tokenizer = AutoTokenizer.from_pretrained('nlptown/bert-base-multilingual-uncased-sentiment')

model = AutoModelForSequenceClassification.from_pretrained('nlptown/bert-base-multilingual-uncased-sentiment')
```

```python
tokens = tokenizer.encode('I Love', return_tensors='pt')
result = model(tokens)
result.logits
int(torch.argmax(result.logits))+1
```

```python
def sentiment_score(tweet):
    tweet = str(tweet)
    tokens = tokenizer.encode(tweet, return_tensors='pt')
    result = model(tokens)
    return int(torch.argmax(result.logits))+1
```

```python
test = df.iloc[0]['Text']
test
```

```python
sentiment_score(test)
```

```python
test = df.iloc[54311]['Text']
test
```

```python
sentiment_score(test)
```

```python
for i in range(length):
    if i % 100 == 0:
        print(i, ' tweets have been processed')
    
    df['sentiment'][i] = sentiment_score(df.iloc[i]['Text'])
```
